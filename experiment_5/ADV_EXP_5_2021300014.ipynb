```r
library(ggplot2)
library(dplyr)
library(wordcloud)
library(RColorBrewer)
```

```r
housing_data <- read.csv("India housing.csv")
housing_data %>% head
```

```r
housing_data <- housing_data %>% na.omit()
```

```r
options(repr.plot.width = 15, repr.plot.height = 7)

location_count = housing_data %>% group_by(Location) %>% summarise(count = n(), .groups = 'drop')

wordcloud(words = location_count$Location, freq = location_count$count, min.freq = 5,
           max.words=75, random.order=FALSE, rot.per=0.35,
           colors=brewer.pal(8, "Dark2"))
```

```r
ggplot(housing_data, aes(x = City, y = Area, fill = City)) +
  geom_boxplot()
```

```r
ggplot(housing_data, aes(x = City, y = Price, fill = City)) +
  geom_violin(bounds = c(0, 5e7))
```

```r
price_area <- ggplot(housing_data, aes(x = Area, y = Price))

price_area +
  geom_point() +
  stat_smooth(method = lm, level=0.95)
```

```r
ggplot(housing_data, aes(Resale, Price)) +
geom_point() + geom_jitter()
```
